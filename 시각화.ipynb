```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
loss_10 = np.load("train_losses_10.npy")
loss_50 = np.load("train_losses_50.npy")
loss_100 = np.load("train_losses_100.npy")
loss_150 = np.load("train_losses_150.npy")
loss_ori = np.load("train_losses_origin.npy")
```

```python
plt.figure(figsize=(10, 6))
plt.plot(loss_10, label="10")
plt.plot(loss_50, label="50")
plt.plot(loss_100, label="100")
plt.plot(loss_150, label="150")
plt.plot(loss_ori, label="origin")
plt.ylim(0, 1)
plt.title("train loss")
plt.legend()
```
