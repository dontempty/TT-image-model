```python
cat_array = np.array(cat_im_list)
dog_array = np.array(dog_im_list)
```

```python
plt.imshow(cat_array[20])
print(256*256*3)
```

```python
cat_tt = tt.tensor_train(cat_array[20], [1, 10, 10, 1])
cat_ttt = tt_to_tensor(cat_tt).astype(np.uint8)
plt.imshow(cat_ttt)
total = 0

for i in range(3):
    m = 1
    for s in cat_tt[i].shape:
        m *= s
    total += m
    
print(total)
print(f"compression ratio: {total/(256*256*3)}")
```
